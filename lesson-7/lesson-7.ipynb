**__6.__ Исследовать на экстремум функцию:**

### $$U=x^2y+\frac{1}{3}y^3+2x^2+3y^2-1$$

Построим график функции:

```python
import pylab
from mpl_toolkits.mplot3d import Axes3D
import numpy
import warnings

warnings.filterwarnings('ignore')

def makeData ():
    x = numpy.arange (-10, 10, 0.1)
    y = numpy.arange (-10, 10, 0.1)
    xgrid, ygrid = numpy.meshgrid(x, y)

    Ugrid = xgrid**2*ygrid+1/3*ygrid**3+2*xgrid**2+3*ygrid**2-1
    return xgrid, ygrid, Ugrid

x, y, U = makeData()

fig = pylab.figure()
axes = Axes3D(fig)

axes.plot_surface(x, y, U)

pylab.show()
```

Приравняем частные производные к 0 и проверим критические точки на экстремумы с помощью критерия Сильвестра.
Сначала решим систему уравнений:

$$\begin{cases}
   U'_{x}=2xy+4x=0 \\
   U'_{y}=x^2+y^2+6y=0 
 \end{cases}$$

```python
import sympy as sym
x,y = sym.symbols('x,y')
eq1 = sym.Eq(2*x*y+4*x,0)
eq2 = sym.Eq(x**2+y**2+6*y,0)
sym.solve([eq1,eq2],(x,y))
```

Построим матрицу Гёссе, т.е. матрицу вида

$$\begin{pmatrix}
U''_{xx} & U''_{xy}\\ 
U''_{yx} & U''_{yy}\\
\end{pmatrix}$$

и вычислим главные миноры по формулам:
$$\Delta_{1}=U''_{xx}$$

$$\Delta_{2}=U''_{xx}\cdot U''_{yy}-(U''_{xy})^2$$

Получим частные производные второго порядка:

$$U''xx = 2y+4$$

$$U''xy=U''yx=2x$$  

$$U''yy=2y+6$$


тогда матрица Гёссе примет вид:

$$\begin{pmatrix}
2y+4 & 2x\\ 
2x & 2y+6\\
\end{pmatrix}$$

Теперь последовательно подставим все корни уравнения в полученную матрицу и посчитаем главные миноры:

Для корней (0, -6) получаем:
$$\Delta_{1}=2y+4=-8<0$$

$$\Delta_{2}=(2y+4)(2y+6)-4x^2=48>0$$

$\boxed{(0,-6)}$ - точка максимума

Для корней (0, 0) получаем:
$$\Delta_{1}=4 > 0$$

$$\Delta_{2}=(2y+4)(2y+6)-4x^2=24>0$$

$$\Delta_{1}=0$$

т.е. в данном случае критерий Сильвестра не работает.
